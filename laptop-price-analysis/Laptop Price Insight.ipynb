```python
import pandas as pd

# Caminho para o arquivo CSV
csv_file = '/content/laptop_price - dataset.csv'

# Carregar o dataset
df = pd.read_csv(csv_file)

# Visualizar as primeiras linhas do DataFrame
print(df.head())

# Verificar a estrutura dos dados
print(df.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizar a distribuição dos preços
sns.histplot(df['Price (Euro)'], bins=30)
plt.title('Distribuição dos Preços dos Laptops')
plt.xlabel('Preço (Euro)')
plt.ylabel('Frequência')
plt.show()
```

Neste exemplo, a faixa de preço de €1000-€1500 tem a maior frequência, significando que há mais laptops com preços nesta faixa em comparação com outras faixas de preço.

Isso nos ajuda a entender não apenas o valor médio ou mediano dos laptops, mas também a distribuição geral dos preços, identificando faixas de preços mais comuns e mais raras.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ajustar o estilo do gráfico
sns.set(style='whitegrid')

# Criar o gráfico de caixa (box plot)
plt.figure(figsize=(12, 6))
sns.boxplot(x='RAM (GB)', y='Price (Euro)', data=df)
plt.title('Relação entre a Quantidade de RAM e o Preço dos Laptops', fontsize=16)
plt.xlabel('RAM (GB)', fontsize=14)
plt.ylabel('Preço (Euro)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

```

A linha dentro de cada caixa mostra a mediana do preço para essa quantidade de RAM.

Se muitos pontos estão longe da linha de referência, isso sugere que o modelo está cometendo erros significativos nas previsões. Esses pontos indicam onde os valores preditos estão muito acima ou abaixo dos valores reais.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar processadores por família/categoria
def categorize_cpu(cpu_type):
    if 'i3' in cpu_type:
        return 'Intel Core i3'
    elif 'i5' in cpu_type:
        return 'Intel Core i5'
    elif 'i7' in cpu_type:
        return 'Intel Core i7'
    elif 'i9' in cpu_type:
        return 'Intel Core i9'
    elif 'Ryzen' in cpu_type:
        return 'AMD Ryzen'
    else:
        return 'Outros'

df['CPU_Category'] = df['CPU_Type'].apply(categorize_cpu)

# Criar o gráfico de caixa (box plot) para os grupos de processadores
plt.figure(figsize=(14, 8))
sns.boxplot(x='CPU_Category', y='Price (Euro)', data=df, palette='viridis')
plt.title('Preço dos Laptops por Categoria de Processador', fontsize=18, weight='bold')
plt.xlabel('Categoria de Processador', fontsize=14)
plt.ylabel('Preço (Euro)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()


```

Estrutura e Interpretação do Gráfico
Eixo X (Categoria de Processador):

No eixo X, temos diferentes categorias de processadores. No nosso exemplo, agrupamos os processadores em categorias amplas como "Intel Core i3", "Intel Core i5", "Intel Core i7", "Intel Core i9", "AMD Ryzen", e "Outros". Cada uma dessas categorias representa uma família de processadores.

Eixo Y (Preço (Euro)):

O eixo Y mostra os preços dos laptops em euros. Este é o valor que estamos tentando analisar e entender como varia de acordo com as diferentes categorias de processadores.

Box Plot (Gráfico de Caixa):

Cada caixa no gráfico representa a distribuição dos preços dos laptops dentro de uma categoria de processadores específica.

Linha do Meio (Mediana): A linha dentro de cada caixa representa a mediana do preço dos laptops naquela categoria de processadores. A mediana é o valor central, onde metade dos laptops tem um preço abaixo e a outra metade tem um preço acima dessa linha.

Caixa (IQR): A caixa em si abrange o intervalo interquartil (IQR), que é o intervalo onde se encontram os 50% centrais dos dados. Isso nos mostra a dispersão dos preços dentro de cada categoria.

Bigodes (Whiskers): Os bigodes se estendem até os valores que estão dentro de 1.5 vezes o IQR a partir dos quartis inferiores (Q1) e superiores (Q3).

Outliers: Pontos fora dos bigodes são outliers, que são preços significativamente mais altos ou mais baixos em comparação com a maioria dos laptops naquela categoria.

Interpretação Específica para Processadores
Intel Core i3, i5, i7, i9:

Esses processadores tendem a mostrar uma tendência crescente de preço conforme a performance e capacidade aumentam de i3 para i9.

Espera-se que laptops com Intel Core i9 tenham preços mais altos em média do que aqueles com Intel Core i3.

AMD Ryzen:

Similarmente, os processadores AMD Ryzen tendem a competir em diferentes faixas de performance e preço, com variantes de alta performance geralmente custando mais.

Outros:

Esta categoria pode incluir processadores menos comuns ou menos potentes, possivelmente com uma maior variação nos preços.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar processadores mais comuns
common_cpus = df['CPU_Type'].value_counts().nlargest(10).index
df_filtered = df[df['CPU_Type'].isin(common_cpus)]

plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# Criar o gráfico de caixa (box plot) com melhorias visuais
sns.boxplot(x='CPU_Type', y='Price (Euro)', data=df_filtered, palette='viridis')
plt.title('Preço dos Laptops por Tipo de Processador (Mais Comuns)', fontsize=18, weight='bold')
plt.xlabel('Tipo de Processador', fontsize=14)
plt.ylabel('Preço (Euro)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

O tamanho da tela também influencia os preços. Laptops com telas maiores tendem a ter preços mais elevados, provavelmente devido à demanda por telas maiores em modelos de alta performance e para tarefas mais exigentes.
